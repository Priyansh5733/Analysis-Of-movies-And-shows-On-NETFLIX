```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("Downloads/Netflix/titles.csv")
# print(df.head())

df.dropna(subset=["type","release_year","age_certification","imdb_score","production_countries","runtime"])

type_counts=df["type"].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_counts.index,type_counts.values,color=["red","blue"])
plt.xlabel("Type")
plt.ylabel("count")
plt.title("Movies VS Shows")
plt.savefig("moviesvstvshows.png")
plt.show()

Agecertification_count=df["age_certification"].value_counts()
plt.figure(figsize=(8,6))
plt.pie(Agecertification_count,labels=Agecertification_count.index,autopct='%1.1f%%')
plt.title("Distribution of Age certification of movies")
plt.savefig("Agecertification.png")
plt.show()

release_counts=df['release_year'].value_counts().sort_index()
plt.scatter(release_counts.index,release_counts.values,color='teal')
plt.xlabel("Year")
plt.ylabel("No. of shows ")
plt.title("showsvsyear")
plt.savefig("Showsvsyear.png")
plt.show()

product_counts=df['production_countries'].value_counts().head(10)
plt.barh(product_counts.index, product_counts.values, color="blue", edgecolor="red")
plt.xlabel("No. of shows")
plt.ylabel("Countries ")
plt.title("Top 10 Show producing Countries")
plt.savefig("Top10countries.png")
plt.show()

content_per_year=df.groupby(['release_year','type']).size().unstack().fillna(0)
fig, ax=plt.subplots(1,2,figsize=(12,5))
ax[0].plot(content_per_year.index,content_per_year['MOVIE'],color='blue')
ax[0].plot(content_per_year.index,content_per_year['SHOW'],color='RED')
plt.suptitle("Comparison of movies and tv shows over the years")
plt.savefig("Comparison.png")
plt.show()

df['genres'] = df['genres'].str.strip("[]").str.replace("'", "")
df['genres'] = df['genres'].str.split(",")
df_exploded = df.explode('genres')
df_exploded['genres'] = df_exploded['genres'].str.strip()

genre_count=df_exploded['genres'].value_counts()
plt.pie(genre_count,labels=genre_count.index,autopct='%1.1f%%')
plt.title("Distribution of Genres of movies")
plt.savefig("Moviesgenres.png")
plt.show()
```
